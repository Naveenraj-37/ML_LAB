```python
# breast_cancer_with_pca_comparison_full.py
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, KFold, cross_val_score, StratifiedKFold
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.base import clone

# Models
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

# Optional libraries
try:
    import xgboost as xgb
    XGB_AVAILABLE = True
except Exception:
    XGB_AVAILABLE = False

try:
    from scipy.stats import ttest_rel
    SCIPY_AVAILABLE = True
except Exception:
    SCIPY_AVAILABLE = False

# ---------- User settings ----------
DATA_PATH = "wdbc.data"            # change if your file is at different path
PCA_VARIANCE = 0.95               # keep 95% explained variance when PCA is used (float) or set integer
TEST_SIZE = 0.20
RANDOM_STATE = 42
CV_FOLDS = 5
# -----------------------------------

# ---------- Load & preprocess ----------
cols = ["ID", "Diagnosis"] + [f"feature_{i}" for i in range(1, 31)]
df = pd.read_csv(DATA_PATH, header=None, names=cols)
df.drop(columns=["ID"], inplace=True)
df["Diagnosis"] = LabelEncoder().fit_transform(df["Diagnosis"])  # M=1, B=0

X_raw = df.drop(columns=["Diagnosis"])
y = df["Diagnosis"].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_raw)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
)

print("\nClass Distribution (0 = Benign, 1 = Malignant):")
print(df["Diagnosis"].value_counts())

plt.figure(figsize=(10, 8))
sns.heatmap(df.drop(columns=["Diagnosis"]).corr(), cmap="coolwarm", cbar=True)
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()

# ---------- Evaluation helper ----------
def evaluate(name, model, X_test, y_test, plot=True):
    """Evaluate model: prints metrics, plots ROC + confusion matrix (if plot=True)."""
    if plot:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
        fig.suptitle(f"{name} Performance", fontsize=16)

    # Predictions and probabilities
    y_pred = model.predict(X_test)
    if hasattr(model, "predict_proba"):
        probs = model.predict_proba(X_test)[:, 1]
    else:
        try:
            dfun = model.decision_function(X_test)
            probs = (dfun - dfun.min()) / (dfun.max() - dfun.min())
        except Exception:
            probs = None

    if plot and probs is not None:
        fpr, tpr, _ = roc_curve(y_test, probs)
        roc_auc = auc(fpr, tpr)
        ax1.plot(fpr, tpr, lw=2, label=f"ROC AUC = {roc_auc:.3f}")
        ax1.plot([0, 1], [0, 1], lw=2, linestyle="--")
        ax1.set_xlim([0.0, 1.0])
        ax1.set_ylim([0.0, 1.05])
        ax1.set_xlabel("False Positive Rate")
        ax1.set_ylabel("True Positive Rate")
        ax1.set_title("ROC Curve")
        ax1.legend(loc="lower right")
    elif plot:
        ax1.text(0.5, 0.5, "No probability scores available", ha="center")
        ax1.set_title("ROC Curve")

    if plot:
        cm = confusion_matrix(y_test, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm)
        disp.plot(ax=ax2, cmap=None)
        ax2.set_title("Confusion Matrix")
        plt.tight_layout()
        plt.show()

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    print(f"\n{name}")
    print("Accuracy :", acc)
    print("Precision:", prec)
    print("Recall   :", rec)
    print("F1 Score :", f1)
    return acc, prec, rec, f1

# ---------- Models + hyperparameter grids ----------
models_params = {
    "Decision Tree": (
        DecisionTreeClassifier(random_state=RANDOM_STATE),
        {"criterion": ["gini", "entropy"], "max_depth": [3, 5, 10, None]}
    ),
    "Random Forest": (
        RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1),
        {"n_estimators": [50, 100, 200], "max_depth": [3, 5, 10, None], "criterion": ["gini", "entropy"]}
    ),
    "AdaBoost": (
        AdaBoostClassifier(random_state=RANDOM_STATE, estimator=DecisionTreeClassifier(random_state=RANDOM_STATE)),
        {"n_estimators": [50, 100, 200], "learning_rate": [0.01, 0.1, 1], "estimator__max_depth": [1, 3, 5]}
    ),
    "Gradient Boosting": (
        GradientBoostingClassifier(random_state=RANDOM_STATE),
        {"n_estimators": [50, 100, 200], "learning_rate": [0.01, 0.1, 0.5], "max_depth": [3, 5, 7]}
    ),
    "KNN": (
        KNeighborsClassifier(),
        {"n_neighbors": [3, 5, 7], "weights": ["uniform", "distance"], "p": [1, 2]}
    ),
    "SVM": (
        SVC(probability=True, random_state=RANDOM_STATE),
        {"kernel": ["rbf", "linear"], "C": [0.1, 1, 10], "gamma": ["scale", "auto"]}
    ),
    "Naive Bayes": (
        GaussianNB(),
        {"var_smoothing": [1e-9, 1e-8, 1e-7]}
    ),
    "Logistic Regression": (
        LogisticRegression(max_iter=5000, solver="saga", random_state=RANDOM_STATE),
        {"C": [0.01, 0.1, 1, 10], "penalty": ["l2", "none"]}
    )
}

if XGB_AVAILABLE:
    models_params["XGBoost"] = (
        xgb.XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=RANDOM_STATE, n_jobs=-1),
        {"n_estimators": [50, 100, 200], "learning_rate": [0.01, 0.1, 0.3], "max_depth": [3, 5, 7], "gamma": [0, 0.1, 0.3]}
    )

# ---------- Results containers ----------
results_table = []   # rows: (name, variant, best_params, acc, prec, rec, f1, elapsed)
trial_tables = {}    # name -> { 'no_pca': df, 'with_pca': df }
best_estimators = {} # name -> {'no_pca': estimator, 'with_pca': estimator}

# ---------- Helper to prefix param grid keys for pipeline ('clf__' prefix) ----------
def prefix_param_grid(param_grid, prefix="clf__"):
    new_grid = {}
    for k, v in param_grid.items():
        new_key = prefix + k
        new_grid[new_key] = v
    return new_grid

# ---------- GridSearch: No-PCA and With-PCA ----------
for name, (model, params) in models_params.items():
    print(f"\n--- Grid Search for {name} ---")
    # Make two pipelines
    pipe_no_pca = Pipeline([("clf", model)])  # using scaled X (X_scaled), so no scaler in pipeline
    pipe_with_pca = Pipeline([("pca", PCA(n_components=PCA_VARIANCE, svd_solver="full")), ("clf", model)])

    # Prefix param grid keys for pipeline
    grid_params = prefix_param_grid(params)  # adds 'clf__' prefix

    # GridSearch No-PCA
    print(" GridSearch (No-PCA)...")
    gs_no = GridSearchCV(pipe_no_pca, param_grid=grid_params, cv=CV_FOLDS, scoring="accuracy", n_jobs=-1, return_train_score=False)
    gs_no.fit(X_train, y_train)
    best_no = gs_no.best_estimator_
    best_estimators.setdefault(name, {})['no_pca'] = best_no
    print("  Best params (No-PCA):", gs_no.best_params_)
    print("  Best CV score (No-PCA):", gs_no.best_score_)

    # Record top-5 trials (no-pca)
    trials_no = []
    for i in range(len(gs_no.cv_results_['params'])):
        trials_no.append({
            **gs_no.cv_results_['params'][i],
            "CV Accuracy": gs_no.cv_results_['mean_test_score'][i]
        })
    trial_df_no = pd.DataFrame(trials_no).sort_values("CV Accuracy", ascending=False).head(5)
    y_pred_no = best_no.predict(X_test)
    trial_df_no["F1 Score (Test)"] = f1_score(y_test, y_pred_no)
    trial_tables.setdefault(name, {})['no_pca'] = trial_df_no

    # Evaluate best_no on test set
    start = time.time()
    best_no.fit(X_train, y_train)
    elapsed = time.time() - start
    res_no = evaluate(f"{name} (No-PCA)", best_no, X_test, y_test, plot=True)
    results_table.append((name, "No-PCA", gs_no.best_params_, *res_no, elapsed))

    # GridSearch With-PCA
    print(" GridSearch (With-PCA)...")
    gs_pca = GridSearchCV(pipe_with_pca, param_grid=grid_params, cv=CV_FOLDS, scoring="accuracy", n_jobs=-1, return_train_score=False)
    gs_pca.fit(X_train, y_train)
    best_pca = gs_pca.best_estimator_
    best_estimators[name]['with_pca'] = best_pca
    print("  Best params (With-PCA):", gs_pca.best_params_)
    print("  Best CV score (With-PCA):", gs_pca.best_score_)

    # Record top-5 trials (with-pca)
    trials_pca = []
    for i in range(len(gs_pca.cv_results_['params'])):
        trials_pca.append({
            **gs_pca.cv_results_['params'][i],
            "CV Accuracy": gs_pca.cv_results_['mean_test_score'][i]
        })
    trial_df_pca = pd.DataFrame(trials_pca).sort_values("CV Accuracy", ascending=False).head(5)
    y_pred_pca = best_pca.predict(X_test)
    trial_df_pca["F1 Score (Test)"] = f1_score(y_test, y_pred_pca)
    trial_tables[name]['with_pca'] = trial_df_pca

    # Evaluate best_pca on test set
    start = time.time()
    best_pca.fit(X_train, y_train)
    elapsed = time.time() - start
    res_pca = evaluate(f"{name} (With-PCA)", best_pca, X_test, y_test, plot=True)
    results_table.append((name, "With-PCA", gs_pca.best_params_, *res_pca, elapsed))

# ---------- Stacking classifiers (3 original variants) ----------
print("\n--- Stacking Classifiers (original single-run variants) ---")
stacking_variants = {
    "Stacking (SVM + NB + DT -> LogisticRegression)": StackingClassifier(
        estimators=[
            ("svm", SVC(probability=True, kernel="rbf", C=1, gamma="scale")),
            ("nb", GaussianNB()),
            ("dt", DecisionTreeClassifier(max_depth=5, random_state=RANDOM_STATE))
        ],
        final_estimator=LogisticRegression(max_iter=500, random_state=RANDOM_STATE),
        n_jobs=-1
    ),
    "Stacking (SVM + NB + DT -> RandomForest)": StackingClassifier(
        estimators=[
            ("svm", SVC(probability=True, kernel="rbf", C=1, gamma="scale")),
            ("nb", GaussianNB()),
            ("dt", DecisionTreeClassifier(max_depth=5, random_state=RANDOM_STATE))
        ],
        final_estimator=RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE),
        n_jobs=-1
    ),
    "Stacking (SVM + DT + KNN -> LogisticRegression)": StackingClassifier(
        estimators=[
            ("svm", SVC(probability=True, kernel="rbf", C=1, gamma="scale")),
            ("dt", DecisionTreeClassifier(max_depth=5, random_state=RANDOM_STATE)),
            ("knn", KNeighborsClassifier(n_neighbors=5))
        ],
        final_estimator=LogisticRegression(max_iter=500, random_state=RANDOM_STATE),
        n_jobs=-1
    )
}

for name, stack_model in stacking_variants.items():
    print(f"\nTraining {name} ...")
    start = time.time()
    stack_model.fit(X_train, y_train)
    elapsed = time.time() - start
    res = evaluate(name, stack_model, X_test, y_test, plot=True)
    results_table.append((name, "Stacking-Default", "Default (base learners tuned separately)", *res, elapsed))
    best_estimators[name] = stack_model

# ---------- NEW: Stacking PCA comparisons ----------
print("\n--- Stacking PCA Comparisons (added) ---")

# 1) Stacking (No-PCA) - same structure as one variant above but explicitly named for comparisons
stack_no_pca = StackingClassifier(
    estimators=[
        ("svm", SVC(probability=True, kernel="rbf", C=1, gamma="scale")),
        ("nb", GaussianNB()),
        ("dt", DecisionTreeClassifier(max_depth=5, random_state=RANDOM_STATE))
    ],
    final_estimator=LogisticRegression(max_iter=500, random_state=RANDOM_STATE),
    n_jobs=-1
)
start = time.time()
stack_no_pca.fit(X_train, y_train)
elapsed = time.time() - start
res_stack_no = evaluate("Stacking (No-PCA)", stack_no_pca, X_test, y_test, plot=True)
results_table.append(("Stacking (SVM+NB+DT)", "No-PCA", "default", *res_stack_no, elapsed))

# 2) Stacking (Global-PCA): pipeline PCA -> Stacking
stack_global_pca_pipeline = Pipeline([
    ("pca", PCA(n_components=PCA_VARIANCE, svd_solver="full")),
    ("stack", StackingClassifier(
        estimators=[
            ("svm", SVC(probability=True, kernel="rbf", C=1, gamma="scale")),
            ("nb", GaussianNB()),
            ("dt", DecisionTreeClassifier(max_depth=5, random_state=RANDOM_STATE))
        ],
        final_estimator=LogisticRegression(max_iter=500, random_state=RANDOM_STATE),
        n_jobs=-1
    ))
])
start = time.time()
stack_global_pca_pipeline.fit(X_train, y_train)
elapsed = time.time() - start
res_stack_global = evaluate("Stacking (Global-PCA)", stack_global_pca_pipeline, X_test, y_test, plot=True)
results_table.append(("Stacking (SVM+NB+DT)", "Global-PCA", "pca_pipeline", *res_stack_global, elapsed))

# 3) Stacking built from best per-model pipelines (best_no_pca)
# choose base model keys that match your best_estimators keys (case sensitive)
base_model_keys = ["SVM", "Naive Bayes", "Decision Tree", "KNN"]  # adapt if you want different set

def safe_get_clone(name, variant):
    try:
        est = best_estimators[name][variant]
        return clone(est)
    except Exception:
        return None

estimators_best_no = []
estimators_best_with = []
for key in base_model_keys:
    p_no = safe_get_clone(key, "no_pca")
    if p_no is not None:
        estimators_best_no.append((key.replace(" ", "_").lower(), p_no))
    p_with = safe_get_clone(key, "with_pca")
    if p_with is not None:
        estimators_best_with.append((key.replace(" ", "_").lower(), p_with))

stack_from_best_no = None
stack_from_best_with = None

if len(estimators_best_no) >= 2:
    stack_from_best_no = StackingClassifier(
        estimators=estimators_best_no,
        final_estimator=LogisticRegression(max_iter=500, random_state=RANDOM_STATE),
        n_jobs=-1
    )
    start = time.time()
    stack_from_best_no.fit(X_train, y_train)
    elapsed = time.time() - start
    res_stack_best_no = evaluate("Stacking (from best_no_pca variants)", stack_from_best_no, X_test, y_test, plot=True)
    results_table.append(("Stacking (from_best)", "from_best_no_pca", "built_from_best_no", *res_stack_best_no, elapsed))
else:
    print("Not enough best_no_pca pipelines found to build stack_from_best_no.")

if len(estimators_best_with) >= 2:
    stack_from_best_with = StackingClassifier(
        estimators=estimators_best_with,
        final_estimator=LogisticRegression(max_iter=500, random_state=RANDOM_STATE),
        n_jobs=-1
    )
    start = time.time()
    stack_from_best_with.fit(X_train, y_train)
    elapsed = time.time() - start
    res_stack_best_with = evaluate("Stacking (from best_with_pca variants)", stack_from_best_with, X_test, y_test, plot=True)
    results_table.append(("Stacking (from_best)", "from_best_with_pca", "built_from_best_with", *res_stack_best_with, elapsed))
else:
    print("Not enough best_with_pca pipelines found to build stack_from_best_with.")

# ---------- 5-Fold Cross-Validation for best estimators ----------
print("\n--- 5-Fold Cross-Validation ---")
kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
cv_results = {}

for model_name, variants in best_estimators.items():
    cv_results[model_name] = {}
    if isinstance(variants, dict):
        for var_name, est in variants.items():
            try:
                scores = cross_val_score(est, X_scaled, y, cv=kf, scoring="accuracy", n_jobs=-1)
                cv_results[model_name][var_name] = scores
                print(f"{model_name} ({var_name}) Fold Accuracies: {np.round(scores,4)}  Avg: {np.mean(scores):.4f}")
            except Exception as e:
                print(f"CV failed for {model_name} ({var_name}): {e}")
    else:
        try:
            scores = cross_val_score(variants, X_scaled, y, cv=kf, scoring="accuracy", n_jobs=-1)
            cv_results[model_name] = scores
            print(f"{model_name} Fold Accuracies: {np.round(scores,4)}  Avg: {np.mean(scores):.4f}")
        except Exception as e:
            print(f"CV failed for {model_name}: {e}")

# Also do paired CV for stacking variants (StratifiedKFold so comparisons are paired)
print("\n--- Paired CV for Stacking Variants (StratifiedKFold) ---")
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
def cv_scores(estimator, Xdata, ydata):
    return cross_val_score(estimator, Xdata, ydata, cv=skf, scoring="accuracy", n_jobs=-1)

scores_stack_no = cv_scores(stack_no_pca, X_scaled, y)
scores_stack_global = cv_scores(stack_global_pca_pipeline, X_scaled, y)
print("Stacking (No-PCA)      Fold Accuracies:", np.round(scores_stack_no,4), " mean:", scores_stack_no.mean(), " std:", scores_stack_no.std())
print("Stacking (Global-PCA)  Fold Accuracies:", np.round(scores_stack_global,4), " mean:", scores_stack_global.mean(), " std:", scores_stack_global.std())

if stack_from_best_no is not None:
    scores_best_no = cv_scores(stack_from_best_no, X_scaled, y)
    print("Stacking (from_best_no) Fold Accuracies:", np.round(scores_best_no,4), " mean:", scores_best_no.mean(), " std:", scores_best_no.std())
else:
    scores_best_no = None

if stack_from_best_with is not None:
    scores_best_with = cv_scores(stack_from_best_with, X_scaled, y)
    print("Stacking (from_best_with) Fold Accuracies:", np.round(scores_best_with,4), " mean:", scores_best_with.mean(), " std:", scores_best_with.std())
else:
    scores_best_with = None

# Paired t-tests (if scipy available)
if SCIPY_AVAILABLE:
    tstat, pval = ttest_rel(scores_stack_no, scores_stack_global)
    print(f"\nPaired t-test (Stack No-PCA vs Global-PCA): t={tstat:.3f}, p={pval:.4f}")
    if pval < 0.05:
        print(" -> Difference is statistically significant (p < 0.05).")
    else:
        print(" -> Difference is NOT statistically significant (p >= 0.05).")
else:
    print("\nscipy not available — skipping paired t-test for stacking comparisons.")

# ---------- Final summary tables ----------
results_df = pd.DataFrame(results_table, columns=[
    "Model", "Variant", "Best_Params", "Accuracy", "Precision", "Recall", "F1", "Elapsed_Sec"
])
print("\n=== Summary results (test set) ===")
display(results_df.sort_values(by="F1", ascending=False))

print("\n=== Top-5 hyperparameter trials (example) ===")
for name, d in trial_tables.items():
    print(f"\n{name} - No-PCA top trials:")
    display(d['no_pca'])
    print(f"\n{name} - With-PCA top trials:")
    display(d['with_pca'])

# Optionally save results to CSV
results_df.to_csv("results_summary.csv", index=False)
for name, d in trial_tables.items():
    d['no_pca'].to_csv(f"{name.replace(' ', '_')}_trials_no_pca.csv", index=False)
    d['with_pca'].to_csv(f"{name.replace(' ', '_')}_trials_with_pca.csv", index=False)

print("\nSaved results_summary.csv and per-model trial CSVs.")
```
